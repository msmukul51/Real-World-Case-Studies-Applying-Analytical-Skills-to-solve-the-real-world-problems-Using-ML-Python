#Dropping Unnecessary columns

```python
df=df.drop(['wind_direction_10_m_above_gnd','wind_direction_900_mb','wind_speed_10_m_above_gnd','wind_speed_900_mb'],axis=1)
```

```python
x=df.drop('generated_power_kw',axis=1)
y=df['generated_power_kw']
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,random_state=42,train_size=20)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Column transformer_1_Scalling
trf_1=ColumnTransformer([('scale',StandardScaler,slice(0,16))])
```

```python
trf_2=XGBRegressor(random_state=42)
```

```python
pipe = Pipeline([
    ('scaler', StandardScaler()),  # Instance of StandardScaler
    ('regressor', XGBRegressor())  # Instance of XGBRegressor
])
```

```python
pipe.fit(x_train,y_train)
```

```python
pipe.score(x_test,y_test)
```
